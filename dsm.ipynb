# 4- start RFM

## 4.1 Recency

```python
df["Order Date"].max()
```

### this is last order

```python
df2=df.groupby("Customer ID")["Order Date"].max()
lastorder=pd.DataFrame(df2)
```

```python
lastorder.reset_index()
lastorder.rename(columns={"Order Date":"Last Order"},inplace=True)
```

```python
lastorder.reset_index(inplace=True)
lastorder
```

```python
merged=pd.merge(df,lastorder,on="Customer ID",how="inner")
```

```python
merged
```

```python
data=merged[["Customer ID","Customer Name","Order Date","Last Order"]]
```

```python
data
```

```python
last_date=data["Order Date"].max()
last_date
```

```python
data.info()
```

```python
data["recency"]=(last_date-data["Last Order"]).dt.days

```

```python
data
```

## 4.2- freq

```python
dataa=data["Customer ID"].value_counts().reset_index()
dataa.rename(columns={"Customer ID":"freq","index":"Customer ID"},inplace=True)
dataa
```

```python
#data.drop(columns=["freq"],inplace=True)
```

```python
data
```

```python
merg=pd.merge(data,dataa,on="Customer ID",how="inner")
merg
```

```python
merg.info()
```

```python
total_sales=df.groupby("Customer ID")["Sales"].sum()
total_sales=pd.DataFrame(total_sales)
total_sales.reset_index(inplace=True)
total_sales
```

```python
data=pd.merge(merg,total_sales,on="Customer ID",how="inner")
data.rename(columns={"Sales":"Montery"},inplace=True)
data
```

```python
data["R_quartile"]=pd.qcut(data["recency"],4,["1","2","3","4"])
```

```python
data["F_quartile"]=pd.qcut(data["freq"],4,["4","3","2","1"])
```

```python
data["M_quartile"]=pd.qcut(data["Montery"],4,["4","3","2","1"])
```

```python
data
```

```python
data["RFM_score"]=data.R_quartile.astype(str)+ data.F_quartile.astype(str) + data.M_quartile.astype(str) 
data
```

```python
special_customers=data[data["RFM_score"]=="111"].sort_values("Montery",ascending=False)
special_customers
```
